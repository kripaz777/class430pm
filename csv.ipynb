```python
data = str()
n = int(input("Enter n = "))
b = open('data.csv','w')
b.write("Name,Age,Address,Phone\n")
for i in range(n):
    name = input("Enter name = ")
    age = int(input("Enter age = "))
    add = input("Enter address = ")
    phone = int(input("Enter phone = "))
    info = f"{name},{age},{add},{phone}\n"
    data = data+info

b.write(data)
b.close()
```

```python
import pandas as pd
data = {}
sn = []
n = int(input("Enter n = "))
for i in range(1,n+1):
    name = input("Enter name = ")
    ntc = int(input("Enter ntc phone no = "))
    ncell = int(input("Enter ncell phone no = "))
    data[name] = [ntc,ncell]
    sn.append(i)
data["Sn"] = sn
df = pd.DataFrame(data)
df.to_csv('datas.csv')
```

```python
import csv   
output = []
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    age = int(input("Enter age = "))
    add = input("Enter address = ")
    info = [name,age,add]
    output.append(info)

f = open('list.csv', 'a')
x = csv.writer(f)
for i in output:
    x.writerow(i)
f.close()
```

```python
import csv
csv_columns = ['Name', 'Age', 'Address']
dict = [
    {'Name':'Aayush','Age':23,'Address':'Katmandu'},
    {'Name':'Shyam','Age':33,'Address':'Patan'}
]
with open("data.csv", 'w') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
    writer.writeheader()
    for data in dict:
        writer.writerow(data)

```
